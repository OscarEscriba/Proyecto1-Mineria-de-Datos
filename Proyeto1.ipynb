##  Variables Cualitativas  
Estas representan categor√≠as o atributos y pueden ser **nominales** o **ordinales**.

````python
#codigo para comprobar si las varaibles cuantitativas siguen una distribucion normal
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import normaltest

# Cargar el dataset 
df = pd.read_csv("movies.csv", encoding="ISO-8859-1") 

# Variables cuantitativas
cuantitativas = [
    "budget", "revenue", "popularity", "voteAvg", "runtime", 
    "voteCount", "genresAmount", "productionCoAmount", 
    "productionCountriesAmount", "actorsAmount", 
    "castWomenAmount", "castMenAmount"
]

# Convertir columnas a valores num√©ricos, forzando errores a NaN (not a number), esto para evitar problemas con el tipo de dato para analizar. 
for col in cuantitativas:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Funci√≥n para revisar la normalidad
def revisar_normalidad(data):
    stat, p_value = normaltest(data)
    if p_value > 0.05:
        return "Distribuci√≥n Normal"
    else:
        return "No Distribuci√≥n Normal"

# Graficar histogramas y revisar normalidad
for var in cuantitativas:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[var].dropna(), kde=True, bins=20)  # Eliminar NaN antes de graficar
    plt.title(f"Distribuci√≥n de {var}")
    plt.show()
    
    resultado_normalidad = revisar_normalidad(df[var].dropna())  # Eliminar NaN antes de la prueba
    print(f"Resultado de normalidad para {var}: {resultado_normalidad}")

````

````python
#¬øLa cantidad de actores influye en los ingresos de las pel√≠culas?¬øse han hecho pel√≠culas con m√°s actores en los √∫ltimos a√±os?
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Cargar el dataset
df = pd.read_csv("movies.csv", encoding="ISO-8859-1")

# Convertir revenue a num√©rico (manejar errores)
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')

# Usar la columna correcta para la cantidad de actores
df['num_actores'] = pd.to_numeric(df['actorsAmount'], errors='coerce')

# Filtrar valores inv√°lidos
df = df[(df['num_actores'] > 0) & (df['revenue'] > 0)]

# Agregar un peque√±o ruido aleatorio para diferenciar puntos con el mismo valor
df['num_actores_jitter'] = df['num_actores'] + np.random.uniform(-0.5, 0.5, df.shape[0])

# Calcular la correlaci√≥n entre n√∫mero de actores e ingresos
correlacion = df[['num_actores', 'revenue']].corr().iloc[0, 1]
print(f"Correlaci√≥n entre cantidad de actores e ingresos: {correlacion:.2f}")

# --- Gr√°fico de dispersi√≥n con escala ajustada ---
plt.figure(figsize=(10, 5))
plt.scatter(df['num_actores_jitter'], df['revenue'], alpha=0.5, color='blue', s=20)
plt.xlabel("Cantidad de Actores")

# Aplicar escala logar√≠tmica a los ingresos si hay valores grandes
plt.yscale('log')

# Ajustar l√≠mites del eje X para mejor visualizaci√≥n
plt.xlim(0, np.percentile(df['num_actores'], 98))  # Recortar extremos superiores

plt.ylabel("Ingresos ($)")
plt.title("Relaci√≥n entre Cantidad de Actores e Ingresos (Escala Ajustada)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print("Se puede obsevar que la correlaci√≥n es 0, por lo que no hay una relaci√≥n clara")
# --- An√°lisis de evoluci√≥n en el tiempo ---
df['releaseDate'] = pd.to_datetime(df['releaseDate'], errors='coerce')
df['year'] = df['releaseDate'].dt.year

# Filtrar datos v√°lidos y calcular el promedio de actores por a√±o
actors_per_year = df.groupby('year')['num_actores'].mean()

# Gr√°fico de evoluci√≥n del n√∫mero de actores por a√±o
plt.figure(figsize=(10, 5))
plt.plot(actors_per_year.index, actors_per_year.values, marker='o', color='green', linestyle='-')
plt.xlabel("A√±o")
plt.ylabel("Cantidad Promedio de Actores")
plt.title("Evoluci√≥n del N√∫mero de Actores en Pel√≠culas")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
print("Hay dos picos de incremento de actores durante los a√±os y este es a principios de los 2000 y en los 2010")


````

````python
#¬øEs posible que la cantidad de hombres y mujeres en el reparto influya en la popularidad y los ingresos de las pel√≠culas?

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Cargar el dataset
df = pd.read_csv("movies.csv", encoding="ISO-8859-1")

# Convertir columnas num√©ricas (manejar errores)
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')
df['popularity'] = pd.to_numeric(df['popularity'], errors='coerce')
df['castWomenAmount'] = pd.to_numeric(df['castWomenAmount'], errors='coerce')
df['castMenAmount'] = pd.to_numeric(df['castMenAmount'], errors='coerce')

# Calcular el total de actores y la proporci√≥n de mujeres en el reparto
df['total_actores'] = df['castWomenAmount'] + df['castMenAmount']
df = df[df['total_actores'] > 0]  # Filtrar pel√≠culas sin reparto

df['proporcion_mujeres'] = df['castWomenAmount'] / df['total_actores']

# Calcular correlaciones
corr_revenue = df[['proporcion_mujeres', 'revenue']].corr().iloc[0, 1]
corr_popularity = df[['proporcion_mujeres', 'popularity']].corr().iloc[0, 1]

print(f"Correlaci√≥n entre proporci√≥n de mujeres en el reparto e ingresos: {corr_revenue:.2f}")
print(f"Correlaci√≥n entre proporci√≥n de mujeres en el reparto y popularidad: {corr_popularity:.2f}")

# --- Gr√°fico: Relaci√≥n entre proporci√≥n de mujeres e ingresos ---
plt.figure(figsize=(10, 5))
plt.scatter(df['proporcion_mujeres'], df['revenue'], alpha=0.5, color='blue', s=20)
plt.xlabel("Proporci√≥n de Mujeres en el Reparto")
plt.ylabel("Ingresos ($)")
plt.yscale('log')  # Escala logar√≠tmica para mejorar visualizaci√≥n
plt.title("Relaci√≥n entre Proporci√≥n de Mujeres en el Reparto e Ingresos")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# --- Gr√°fico: Relaci√≥n entre proporci√≥n de mujeres y popularidad ---
plt.figure(figsize=(10, 5))
plt.scatter(df['proporcion_mujeres'], df['popularity'], alpha=0.5, color='red', s=20)
plt.xlabel("Proporci√≥n de Mujeres en el Reparto")
plt.ylabel("Popularidad")
plt.title("Relaci√≥n entre Proporci√≥n de Mujeres en el Reparto y Popularidad")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

````

Como podemos observar, en el primer gr√°fico la correlaci√≥n es negativa, por lo que con m√°s mujeres en el reparto se tiene a tener menor ingresos. En el segundo gr√°fico la correlacion es positiva, por lo que entre m√°s mujeres en el reparto mayor ser√° la popularidad.

````python
#¬øC√≥mo se correlacionan los presupuestos con los ingresos? ¬øLos altos presupuestos significan altos ingresos? Haga los gr√°ficos que necesite, histograma, diagrama de dispersi√≥n

import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv("movies.csv", encoding="ISO-8859-1")

# Convertir las columnas de presupuesto e ingresos a num√©rico
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')

# Eliminar valores nulos en presupuesto e ingresos
df_clean = df.dropna(subset=['budget', 'revenue'])

# Calcular la correlaci√≥n entre presupuesto e ingresos
correlation = df_clean[['budget', 'revenue']].corr().iloc[0, 1]

# Crear gr√°ficos
plt.figure(figsize=(12, 5))

# Histograma del presupuesto
plt.subplot(1, 2, 1)
sns.histplot(df_clean['budget'], bins=30, kde=True, color='blue')
plt.xlabel("Presupuesto ($)")
plt.ylabel("Frecuencia")
plt.title("Distribuci√≥n de los Presupuestos")

# Histograma de los ingresos
plt.subplot(1, 2, 2)
sns.histplot(df_clean['revenue'], bins=30, kde=True, color='green')
plt.xlabel("Ingresos ($)")
plt.ylabel("Frecuencia")
plt.title("Distribuci√≥n de los Ingresos")

plt.tight_layout()
plt.show()

# Diagrama de dispersi√≥n
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df_clean['budget'], y=df_clean['revenue'], alpha=0.5, color='red')
plt.xlabel("Presupuesto ($)")
plt.ylabel("Ingresos ($)")
plt.title(f"Relaci√≥n entre Presupuesto e Ingresos (Correlaci√≥n: {correlation:.2f})")
plt.xscale('log')
plt.yscale('log')  # Escala logar√≠tmica para mejor visualizaci√≥n
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print(f"Correlaci√≥n entre presupuesto e ingresos: {correlation:.2f}")
````

Un presupuesto alto no garantiza que la pel√≠cula tendra altos ingresos, otros factores como marketing, calidad y demanda influenciar√°n. Aunque si es observable que puede que exista un relaci√≥n entre un presupuesto alto y un ingreso alto.

````python
#¬øSe asocian ciertos meses de lanzamiento con mejores ingresos?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv("movies.csv", encoding="ISO-8859-1")

# Convertir la fecha de lanzamiento a formato datetime
df['release_date'] = pd.to_datetime(df['releaseDate'], errors='coerce')

# Extraer el mes de lanzamiento
df['release_month'] = df['release_date'].dt.month

# Calcular ingresos promedio por mes
monthly_revenue = df.groupby('release_month')['revenue'].mean()

# Crear gr√°fico de ingresos por mes
plt.figure(figsize=(10, 5))
sns.barplot(
    x=monthly_revenue.index, 
    y=monthly_revenue.values, 
    palette="viridis", 
    hue=monthly_revenue.index, 
    dodge=False,
    legend=False
)
plt.xlabel("Mes de Lanzamiento")
plt.ylabel("Ingresos Promedio ($)")
plt.title("Ingresos Promedio por Mes de Lanzamiento")
plt.xticks(ticks=range(12), labels=[
    "Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", 
    "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"
], rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

# Identificar el mes con mayores ingresos promedio
best_month = monthly_revenue.idxmax()
print(f"El mes con mayores ingresos promedio es: {best_month}")


````

````python
#En qu√© meses se han visto los lanzamientos con mejores ingresos? ¬øcuantas pel√≠culas, en promedio, se han lanzado por mes?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Asegurarnos de que la columna release_date est√© en formato datetime
df['releaseDate'] = pd.to_datetime(df['releaseDate'], errors='coerce')

# Filtrar datos v√°lidos para evitar errores
valid_data = df.dropna(subset=['releaseDate', 'revenue'])

# Obtener el mes de los lanzamientos con mayores ingresos
max_revenue_months = valid_data.loc[valid_data['revenue'] == valid_data['revenue'].max(), 'release_date']
top_months = max_revenue_months.dt.month.unique()

# Lista de nombres de meses
meses_top = [
    "Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", 
    "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"
]

# Mostrar nombres de los meses con mejores ingresos
top_month_names = [meses_top[month - 1] for month in top_months if not pd.isna(month)]
print(f"Mes(es) con los lanzamientos con mejores ingresos: {top_month_names}")

# Calcular el n√∫mero de pel√≠culas lanzadas por mes
valid_data['release_month'] = valid_data['release_date'].dt.month
monthly_counts = valid_data.groupby('release_month')['release_date'].count()

# Promedio de pel√≠culas lanzadas por mes
average_movies_per_month = monthly_counts.mean()
print(f"En promedio, se han lanzado aproximadamente {average_movies_per_month:.2f} pel√≠culas por mes.")
````

````python
#¬øC√≥mo se correlacionan las calificaciones con el √©xito comercial?

import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar datos v√°lidos para evitar errores
valid_data = df.dropna(subset=['revenue', 'voteAvg'])

# Calcular la correlaci√≥n entre calificaciones y √©xito comercial
correlation = valid_data[['voteAvg', 'revenue']].corr().loc['voteAvg', 'revenue']
print(f"La correlaci√≥n entre las calificaciones y los ingresos es: {correlation:.2f}")

# Visualizar la relaci√≥n con un gr√°fico de dispersi√≥n
plt.figure(figsize=(10, 5))
sns.scatterplot(data=valid_data, x='voteAvg', y='revenue', alpha=0.6, color='purple')
plt.title("Relaci√≥n entre Calificaciones y √âxito Comercial")
plt.xlabel("Calificaci√≥n Promedio")
plt.ylabel("Ingresos ($)")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
````

Dada la correlaci√≥n podemos decir que si existe una relacion entre la calificaci√≥n y los ingresos de una pel√≠cula, siendo que entre mejor puntuada este, mayores fueron las ganancias.

````python
#¬øQu√© estrategias de marketing, como videos promocionales o p√°ginas oficiales, generan mejores resultados?

# Crear una nueva columna para identificar si la pel√≠cula tiene p√°gina oficial y video promocional
df['has_homepage'] = df['homePage'].notna().astype(int)
df['has_video'] = df['video'].astype(bool).astype(int)

# Calcular los ingresos promedio seg√∫n la estrategia de marketing
marketing_results = df.groupby(['has_homepage', 'has_video'])['revenue'].mean().reset_index()

# Visualizar los resultados
plt.figure(figsize=(10, 5))
sns.barplot(data=marketing_results, x='has_homepage', y='revenue', hue='has_video', palette='magma')
plt.title("Ingresos promedio seg√∫n estrategias de marketing")
plt.xlabel("Tiene P√°gina Oficial (0 = No, 1 = S√≠)")
plt.ylabel("Ingresos Promedio ($)")
plt.legend(title="Video Promocional")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
````

Se puede observar que si una pelicula tiene p√°gina oficial y video promocional, tiende a tener mejores ingresos.
