- **`productionCountry`** ‚Üí Pa√≠s de producci√≥n.

````python
#codigo para comprobar si las varaibles cuantitativas siguen una distribucion normal
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import normaltest

# Cargar el dataset 
df = pd.read_csv("movies.csv", encoding="ISO-8859-1") 

# Variables cuantitativas
cuantitativas = [
    "budget", "revenue", "popularity", "voteAvg", "runtime", 
    "voteCount", "genresAmount", "productionCoAmount", 
    "productionCountriesAmount", "actorsAmount", 
    "castWomenAmount", "castMenAmount"
]

# Convertir columnas a valores num√©ricos, forzando errores a NaN (not a number), esto para evitar problemas con el tipo de dato para analizar. 
for col in cuantitativas:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Funci√≥n para revisar la normalidad
def revisar_normalidad(data):
    stat, p_value = normaltest(data)
    if p_value > 0.05:
        return "Distribuci√≥n Normal"
    else:
        return "No Distribuci√≥n Normal"

# Graficar histogramas y revisar normalidad
for var in cuantitativas:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[var].dropna(), kde=True, bins=20)  # Eliminar NaN antes de graficar
    plt.title(f"Distribuci√≥n de {var}")
    plt.show()
    
    resultado_normalidad = revisar_normalidad(df[var].dropna())  # Eliminar NaN antes de la prueba
    print(f"Resultado de normalidad para {var}: {resultado_normalidad}")

````

````python
#cuantas peliculas se hicieron en cada anio, y en que anio se hicieron mas peliculas
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv("movies.csv", encoding="ISO-8859-1")

# Convertir la columna 'releaseDate' a formato datetime para extraer el a√±o
df['releaseDate'] = pd.to_datetime(df['releaseDate'], errors='coerce')  # Maneja errores
df['year'] = df['releaseDate'].dt.year  # Extraer solo el a√±o

# Contar cu√°ntas pel√≠culas se hicieron por a√±o
movies_per_year = df['year'].value_counts().sort_index()

# Encontrar el a√±o con m√°s pel√≠culas
max_year = movies_per_year.idxmax()
max_movies = movies_per_year.max()

print(f"El a√±o con m√°s pel√≠culas fue {max_year} con {max_movies} pel√≠culas.")

# Graficar los resultados
plt.figure(figsize=(12, 6))
movies_per_year.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Cantidad de pel√≠culas por a√±o")
plt.xlabel("A√±o")
plt.ylabel("Cantidad de pel√≠culas")
plt.xticks(rotation=90)  # Rotar etiquetas para mejor visualizaci√≥n
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv("movies.csv", encoding="ISO-8859-1")

# Convertir la columna 'releaseDate' a formato datetime
df['releaseDate'] = pd.to_datetime(df['releaseDate'], errors='coerce')  # Maneja errores

# Extraer el g√©nero principal (primer g√©nero de la lista separada por "|")
df['mainGenre'] = df['genres'].str.split('|').str[0]

# --- 1. G√©nero principal de las 20 pel√≠culas m√°s recientes ---
top_recent_movies = df.sort_values(by='releaseDate', ascending=False).head(20)
print("G√©nero principal de las 20 pel√≠culas m√°s recientes:")
print(top_recent_movies[['title', 'mainGenre']])

# --- 2. G√©nero principal que predomina en el conjunto de datos ---
genre_counts = df['mainGenre'].value_counts()
top_genre = genre_counts.idxmax()  # G√©nero m√°s frecuente
top_genre_count = genre_counts.max()
print(f"\nEl g√©nero principal que predomina en el conjunto de datos es '{top_genre}' con {top_genre_count} pel√≠culas.")

# Gr√°fico de barras del conteo de g√©neros principales
plt.figure(figsize=(12, 6))
genre_counts.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title("Distribuci√≥n de G√©neros Principales en el Conjunto de Datos")
plt.xlabel("G√©nero")
plt.ylabel("Cantidad de Pel√≠culas")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- 3. G√©nero principal de las pel√≠culas m√°s largas ---
longest_movies = df.sort_values(by='runtime', ascending=False).head(10)  # 10 pel√≠culas m√°s largas
print("\nG√©nero principal de las pel√≠culas m√°s largas:")
print(longest_movies[['title', 'runtime', 'mainGenre']])
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv("movies.csv", encoding="ISO-8859-1")


# Extraer el g√©nero principal (primer g√©nero de la lista separada por "|")
df['mainGenre'] = df['genres'].str.split('|').str[0]

# ---  G√©nero principal que obtuvo mayores ganancias ---
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')  # Convertir a num√©rico, manejar errores

# Calcular las ganancias totales por g√©nero principal
revenue_by_genre = df.groupby('mainGenre')['revenue'].sum().sort_values(ascending=False)

# Encontrar el g√©nero con mayores ganancias
top_revenue_genre = revenue_by_genre.idxmax()
top_revenue_amount = revenue_by_genre.max()
print(f"\nEl g√©nero principal que obtuvo mayores ganancias es '{top_revenue_genre}' con un total de {top_revenue_amount} en ganancias.")

# Gr√°fico de barras de las ganancias por g√©nero principal
plt.figure(figsize=(12, 6))
revenue_by_genre.plot(kind='bar', color='lightblue', edgecolor='black')
plt.title("Ganancias Totales por G√©nero Principal")
plt.xlabel("G√©nero")
plt.ylabel("Ganancias Totales")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````
